```python
import numpy as np
import pandas as pd
```

```python
top_left = pd.read_csv('top1-25-1.csv')
top_right = pd.read_csv('top1-25-2.csv')
bottom_left = pd.read_csv('top26-50-1.csv')
bottom_right = pd.read_csv('top26-50-2.csv')
```

---
Glue together one table from 4 fragments:

⬇ ⬇ ⬇ ⬇ ⬇

```python
bottom = pd.concat([bottom_left, bottom_right], axis=1)
top = pd.merge(top_left, top_right)

df = pd.concat([top, bottom])
```

---
# Arrange the index - make it start at 1.



```python
df.reset_index(drop=True, inplace=True)
df.index = df.index + 1
df
```

---
# Create an object for grouping by genre.




```python
df_grouped_genre = df.groupby('Genre').size()
```

---
# Which genres occur more than 3 times in the table?



```python
df_grouped_genre[df_grouped_genre > 3]
```

---
# What genre is the most popular? What is the least popular?



```python
grouped = df.groupby('Genre')["Popularity"].mean()
most_popular_genre = grouped.idxmax()
highest_popularity_score = grouped.max()

least_popular_genre = grouped.idxmin()
lowest_popularity_score = grouped.min()
```

```python
print(f'{most_popular_genre} - {highest_popularity_score}')
print(f'{least_popular_genre} - {lowest_popularity_score}')
```

---
# Create a table showing which genre has the highest average for each indicator and the average itself.

```python
excluded = ['Track.Name', 'Artist.Name', 'Genre']
indicators = [col for col in df.columns if col not in excluded]

result = []

for indicator in indicators:
    averages = df.groupby('Genre')[indicator].mean()

    max_genre = averages.idxmax()
    max_value = averages.max()

    result.append([indicator, max_genre, max_value])

result_df = pd.DataFrame(result, columns=['Indicator', 'Genre', 'Average'])

print(result_df)
```

# Let's go back to the pasted table. Extract all rows that have NaN values.

```python
rows = df.loc[df.isnull().any(axis=1)]
rows
```

---
# In the 'Genre' column, change the NaN values ​​to 'pop'

```python
rows['Genre'].fillna('Pop', inplace=True)
rows
```

---
# In the 'Popularity' column, replace the missing values ​​with the column mean

```python
rows['Popularity'].fillna(value=rows['Popularity'].mean(), inplace=True)
rows
```

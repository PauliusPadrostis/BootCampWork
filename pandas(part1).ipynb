---
Kokie miestai 1897 turėjo daugiau gyventojų, negu 2019?

⬇ ⬇ ⬇ ⬇ ⬇

```python
df[df['1897'] > df['2019']]
```

---
Kuriuose miestuose padaugėjo gyventojų nuo 2011 iki 2019?

⬇ ⬇ ⬇ ⬇ ⬇

```python
df[df['2011'] < df['2019']]
```

---
Kuriuose miestuose gyventojų skaičius nuosekliai mažėjo nuo pat 1897m.?

⬇ ⬇ ⬇ ⬇ ⬇

```python
df[
    (df['1897'] > df['1923']) &
    (df['1923'] > df['1959']) &
    (df['1959'] > df['1970']) &
    (df['1970'] > df['1979']) &
    (df['1979'] > df['1989']) &
    (df['1989'] > df['2001']) &
    (df['2001'] > df['2011']) &
    (df['2011'] > df['2019'])
]
```

---
Suraskite labiausiai procentaliai gyventojų skaičiumi padidėjusį ir sumažėjusį miestus nuo 1989m.

⬇ ⬇ ⬇ ⬇ ⬇

```python
df['percentage_change'] = ((df['2019'] - df['1989']) / df['1989']) * 100
```

```python
df['percentage_change'].idxmax(), df['percentage_change'].max()
```

```python
df['percentage_change'].idxmin(), df['percentage_change'].min()
```

# Write a function that translates the name of the month into a number. Use .apply() to replace the month names with numbers in the table.

```python
def trans_month(df_months):

  months = {1: 'January',
            2: 'February',
            3: 'March',
            4: 'April',
            5: 'May',
            6: 'June',
            7: 'July',
            8: 'August',
            9: 'September',
            10: 'October',
            11: 'November',
            12: 'December'}


  for month in months:
    if months[month] == df_months:
      return month


df['month'] = df['month'].apply(trans_month)
df.head()
```

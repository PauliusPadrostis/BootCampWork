Defining train_test_split

```python
x = fish[['Length1', 'Length2', 'Length3', 'Height', 'Width']]
```

```python
y = fish['Weight']
```

```python
x, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
```

TRAINING

using np.log to account for negative values in predictions

```python
y_train_transformed = np.log(y_train + 1)
```

```python
reg = LinearRegression().fit(x_train, y_train_transformed)
```

TESTING

```python
prediction = reg.predict(x_test)
```

```python
predictions = np.exp(prediction) - 1
```

due to using log on y_train, wont be able to calculate accuracy score with just .score. Need r2_score.

```python
from sklearn.metrics import r2_score
```

```python
predictions
```

```python
pred = pd.Series(data=predictions, name='Guess')
```

```python
results = pd.concat([y_test.reset_index(), pred], axis=1)[['Weight', 'Guess']]
```

```python
results
```

```python
reg.score(x_test, y_test)
```

```python
r2_score(y_test, predictions)
```
